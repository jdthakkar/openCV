```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
building_img=cv2.imread('building-windows.jpg',0)
r,img=cv2.threshold(building_img,100,255,cv2.THRESH_BINARY)
plt.imshow(img,cmap="gray")
```

```python
img=cv2.imread('Piano_Sheet_Music.png',0)
adaptive_img=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,7)
plt.imshow(adaptive_img,cmap="gray")
```

```python
img_circle=cv2.imread('circle.jpg',0)
img_rect=cv2.imread('rectangle.jpg',0)
```

```python
img_1=cv2.bitwise_and(img_circle,img_rect,mask=None)
plt.imshow(img_1,cmap='grey')
```

```python
img_1=cv2.bitwise_or(img_circle,img_rect,mask=None)
plt.imshow(img_1,cmap='grey')
```

```python
img_1=cv2.bitwise_xor(img_circle,img_rect,mask=None)
plt.imshow(img_1,cmap='grey')
```

```python
img_1=cv2.bitwise_not(img_circle,img_rect,mask=None)
plt.imshow(img_1,cmap='grey')
```
