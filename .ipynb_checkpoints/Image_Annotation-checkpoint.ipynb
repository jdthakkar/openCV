```python
import cv2
import matplotlib.pyplot as plt
```

```python
og_img=cv2.imread('img_3.jpg',1)
plt.imshow(og_img)
```

```python
cv2.line(og_img, (200, 100), (400, 100), (0, 255, 255), thickness=2, lineType=cv2.LINE_AA);
plt.imshow(og_img)
```

```python
cv2.rectangle(og_img, (450, 50), (700, 700), (12, 255, 255), thickness=-1, lineType=cv2.LINE_AA);
plt.imshow(og_img)
```

```python
cv2.circle(og_img, (950, 620), 100, (122, 205, 5), thickness=2, lineType=cv2.LINE_AA);
plt.imshow(og_img)
```

```python
cv2.putText(og_img,"Just Learning",(200,600),cv2.FONT_HERSHEY_COMPLEX,-2,(55,77,223))
plt.imshow(og_img)
```
