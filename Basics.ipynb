```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('img_1.jpg',1)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
cv2.namedWindow('W_1')
cv2.imshow('W_1',img)
cv2.waitKey(8000)
cv2.destroyWindow('W_1')
```
