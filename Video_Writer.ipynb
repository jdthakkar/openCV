```python
import cv2
import matplotlib.pyplot as plt
```

```python
vid=cv2.VideoCapture('race_car.mp4')
```

```python
if not vid.isOpened:
    print("The Video Cannot be Opened")
else:
    stat,frame=vid.read()
    if stat:
        plt.imshow(frame)
    else:
        print("The Video Cannot be Opened")      
```

```python
width=int(vid.get(3))
height=int(vid.get(4))
avi_writer=cv2.VideoWriter("t1.avi",cv2.VideoWriter_fourcc('M','J','P','G'),10,(width,height))
mp4_writer=cv2.VideoWriter("t2.mp4",cv2.VideoWriter_fourcc(*"XVID"),10,(width,height))
```

```python
while vid.isOpened():
    stat,frame=vid.read()
    if stat:
        avi_writer.write(frame)
        mp4_writer.write(frame)
    else:
        break;

vid.release()
avi_writer.release()
mp4_writer.release()
```
